```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
fig = plt.figure()
```

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(x, y);
```

# Standard Example

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(x, y, 'b--', linewidth = 3, label = 'sine');
ax.plot(x, np.cos(x), 'g-.', linewidth = 3, label = 'cosine');
ax.set(xlim=(0, 15), ylim=(-2, 2), xlabel = 'x', ylabel = 'y')
ax.legend();
```

# Subplot

```python
fig, axes = plt.subplots(3, 2, figsize = (8, 8))

axes[0, 0].plot(x, y);
axes[1, 1].plot(x, np.cos(x), 'g');
axes[2, 1].plot(np.tan(x), y, 'r');
```
