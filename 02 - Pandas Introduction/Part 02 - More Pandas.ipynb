```python
def remove_symbol(x):
    return x.replace("FDA_", "")

df.vitamins = df.vitamins.apply(remove_symbol)
df.head()
```

```python
def normalisation(x):
    return x - x.mean()

df.rating = df.rating.transform(normalisation)
df.head()
```

```python
mapper = {
    "C": 0,
    "H": 1
}

df.type = df.type.map(mapper)
df.head()
```
